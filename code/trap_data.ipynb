```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
import statsmodels.graphics.regressionplots as smg
```

```python
plt.rcParams['savefig.dpi'] = 600
plt.rcParams['figure.figsize'] = (6,3.5)
```

```python
def fit(data):
    """Return a linear model of rate to force."""
    model = smf.ols("rate ~ force", data=data).fit()
    return model
    
def fitandplot(data, name='Linear fit'):
    model = fit(data)
    plt.figure()
    ax = plt.subplot(111)
    plt.scatter(data['force'], data['rate'], s=50)
    plt.ylabel('Rate (1/s)')
    plt.xlabel('Force (pN)')
    plt.title(name)
    plt.figtext(.15, .65, name + '\n' + str(model.params), size='x-large')
    _ = smg.abline_plot(model_results=model, color='k', linestyle='--', ax=ax)
    return model

def summary(filename, folder='trap_data/', direction='plus'):
    if direction == 'minus':
        sign = 1
    else:
        sign = -1
    data = pd.read_csv(folder + filename)
    model = fitandplot(data[sign * data['force'] > 0], name=filename[:-4])
    return model
```

```python
def getfits(direction):
    files = !ls trap_data/
    files = [f for f in files if f.endswith('.csv')]
    fits = pd.DataFrame()
    for f in files:
        model = summary(f, folder='trap_data/', direction=direction)
        fits[f[:-8]] = model.params
    fits = fits.T
    tensionpernm = 1/float(12)
    fits['Slope'] = fits['force'] * tensionpernm
    return fits
```
