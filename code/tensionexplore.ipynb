```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import run_simulation as rs
```

```python
resultdict = {}
for i in xrange(4,17, 4):
    resultdict[i] = rs.run_simulation(num_traces=200, nmperpn=i, time_step=0.001)
```

```python
compdict = {}
for k, v in resultdict.iteritems():
    compdict[k] = rs.compare_results(v)
```

```python
plt.rcParams['savefig.dpi'] = 300
plt.rcParams['figure.figsize'] = [4,16]
```

```python
i = 1
for k, comp in sorted(compdict.items(), key=lambda x: x[0]):
    plt.subplot(410 + i)
    i += 1
    exMean = comp['exMean']
    simMean = comp['simMean']
    plt.bar(np.arange(.8, 6), simMean, align='center', width=.4, 
            color='purple', yerr=comp['simS.D.'], ecolor='k', label='Model')
    plt.bar(np.arange(1.2, 7), exMean, align='center', width=.4, 
            color='yellow', yerr=comp['exS.D.'], ecolor='k', label='Experiment')
    plt.ylabel(r'Speed $s^{-1}$', size='x-large')
    if i == 5:
        plt.xlabel('Mutant')
        plt.xticks(xrange(1,7), comp.index, rotation=90, size='large')
    plt.legend()
    plt.title('%s nm per pN of tension' % k);
plt.savefig('../output/tensionexplore.eps')
```
